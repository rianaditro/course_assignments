#### Exploring Data

```python
"""
Execute this cell before continue
""" 

import numpy as np
import pandas as pd

df = pd.DataFrame.from_records((
    (2, 83.82, 8.4),
    (4, 99.31, 16.97),
    (3, 96.52, 14.41),
    (6, 114.3, 20.14),
    (4, 101.6, 16.91),
    (2, 86.36, 12.64),
    (3, 92.71, 14.23),
    (2, 85.09, 11.11),
    (2, 85.85, 14.18),
    (5, 106.68, 20.01),
    (4, 99.06, 13.17),
    (5, 109.22, 15.36),
    (4, 100.84, 14.78),
    (6, 115.06, 20.06),
    (2, 84.07, 10.02),
    (7, 121.67, 28.4),
    (3, 94.49, 14.05),
    (6, 116.59, 17.55),
    (7, 121.92, 22.96),
), columns=("age", "height_cm", "weight_kg"))

df
```

```python
"""
Understand data structure
"""

df.head()         # Show first 5 rows
df.tail()         # Show last 5 rows
df.info()         # Column types and missing values
df.shape          # Rows and columns
df.columns        # List of column names
```

```python
"""
Summary statistics
"""

df.describe()     # Mean, std, min, 25%, 50%, 75%, max
df.nunique()      # Count of unique values
```

```python
"""
Check data quality
"""

df.isnull().sum()
```

```python
"""
Grouping
""" 

df.groupby(by="age").mean()
```

```python
"""
Aggregate
""" 

df.groupby(by='age').agg({
    'height_cm': 'mean',
    'weight_kg': ['mean','max', 'min'],
})
```
